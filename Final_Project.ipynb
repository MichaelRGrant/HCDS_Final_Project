### Introduction

Organic foods have become increasingly popular in recent years. People are becoming more aware that the food they eat has a greater impact on their health, and there has been a growing movement to reduce the amount of chemicals that we ingest. First-world living has many perks, but the first-world cancer rates are much higher than anywhere else on the planet. One possible reason for this is the increase use of chemical pesticides on our food, although this is a just speculation. Since the Green Revolution, conventional agriculture has relied heavily on the use of these chemicals. This did have the intended effect of increasing yields, but the long-term effects of ingesting these products is far more difficult to ascertain. 

The process to get a pesticide approved for agricultural use is much easier in terms of the amount of time it takes to determine that it has deleterious effects for humans. One agrochemical company that often takes the most heat, Monsanto, is responsible for the development of many now banned substances including dichloro-diphenyl-trichloroethane or DDT, 2-4 D an ingredient in the now infamous Agent Orange, a plant defoliating used in the Vietnam War, polychlorinated biphenyls or PCBs a now known carcinogen, and many more. As more of these chemicals are found to be extremely harmful, there is a greater desire for our food to be chemical free and hence the increase in demand for organic foods.

Organic foods are grown without the use of non-approved pesticides or fertilizers. Organic food is generally more difficult to produce on the same scale as conventional agriculture and the amount of loss to disease is greater resulting in reduced overall yields [1]. This fact inevitably drives the cost up compared to conventionally grown varieties. Nevertheless, it seems that most consumers buy organic foods for reasons of food safety [2, 3], knowing that pesticides were not present [4-6] or that the food is perceived to taste better. 

These consumers understand that in order to get the things they want from their food, they will have to pay a premium. There is a large variation in what people are willing to spend to get these advantages. In the US Jolly, Schutz, Diaz-Knauf and Johal [3] found that consumers were willing to pay a 37% price premium while in New York some consumers were willing to pay as much as 100% more [2]. Overall, Yiridoe, Bonti-Ankomah and Martin [1] reported that consumers are willing to pay a 10 – 20% premium above the same food conventionally grown, with premiums above 20% showing a sharp decline in demand. 

In this study, I wanted to investigate what the price difference was for vegetables and fruits over time. My overarching hypothesis is that the price differential between organic and conventional products would not, on average, be greater than 20%. Furthermore, I would like to investigate if demand for these products is increasing. By looking at the prices over time, this question can be answered. Overall, I wish to explore how the price of organic products changes throughout the year, if at all, what the price differential is on different products, and if there is a difference in locations where these products are sold. It will also be interesting to identify which products have the highest premium on organic versions, and what the demand is like on those products.

In order to streamline the cleaning process, two functions were written and used for the appropriate datasets. The data was not all the same when subsetted into the different groups. The Archived Tables and the Current Tables downloaded from the website came in different forms and required unique cleaning operations.

The following function cleans data that was downloaded from the Archived Tables on the website linked to above. The function takes data in the form shown above and transforms it into a form that can be aggregated into a larger dataframe for analysis. 

```r
single_year_clean <- function(data, year){
    
    ###
    # This function takes data, cleans and reforms it into a tidy form to be used for further analysis.
    #
    # param: data - a dataframe
    # param: year - a four digit numerical representation of the year
    ###
    
    row.names(data) <- NULL # reorder the row names starting at 1
    colnames(data) = as.character(unlist(data[1, ]))
    data <- data[-c(1,2), ] # drop the first two rows
    names(data) <- c('year', 'food_type', 'month', 'organic', 'conventional')  # rename columns
    # remove empty rows
    data <- data[rowSums(is.na(data)) != ncol(data),]
    
    # add a column denoting the location the data was being collected for
    row.names(data) <- NULL
    boston_idx <- which(data$food_type == 'Boston')
    sf_idx <- which(data$food_type == 'San Francisco')
    data[1:sf_idx-1, 'location'] = 'Boston'
    data[sf_idx:nrow(data), 'location'] = 'San Francisco'
    
    # remove the last four rows of metadata which is not needed
    row.names(data) <- NULL
    data <- data[-c(boston_idx, sf_idx), ]
    row.names(data) <- NULL
    num_row <- nrow(data)
    data <- data[-c((num_row-3):num_row), ]
    
    # clean the date information in the table
    cols = c('year', 'month')
    data$year <- year
    data$month <- sapply(data$month, 
                         function(x) ifelse(x=='Sept', 'Sep', as.character(x)))
    data$month <- sapply(data$month,
                         function(x) ifelse(x=='Ded', 'Dec', as.character(x)))
    data$date <- apply(data[, cols], 1, paste, collapse = '-')
    data <- data[ , !( names( data ) %in% cols ) ]
    data$date <- (paste0(data$date, "-01"))
    
    data$date <- gsub(pattern = " ", replacement = "", x = data$date)
    
    data$date <- strptime(data$date, format = '%Y-%b-%d')
    data <- apply(data, 2, function(x) ifelse(x=='--', NA, x))
    data <- apply(data, 2, function(x) gsub(pattern = "\\*", replacement = "", x))
    data <- data.frame(data)
    
    # create the tidy dataframe
    data <- gather(data, key = 'type', value = 'price', 2:3)
    data <- transform(data, date = as.Date(date, format = '%Y-%m-%d'),
                      type = factor(type),
                      price = as.numeric(price))    
    return(data)
}
```

Because this data is in a very different form, a seperate function was needed to be written to streamline the cleaning process. The following function was created for this data form. This function takes the above data and forms it similar to the way the earlier function does. 

```r
vegetables_clean <- function(data){
    
    ###
    # This function takes data, cleans and reforms it into a tidy form to be used for further analysis.
    #
    # param: data - a dataframe
    ###
    
    colnames(data) = as.character(unlist(data[1, ]))
    data <- data[-c(1,2), ]
    
    data <- apply(data, 2, function(x) ifelse(x=='N\\A', NA, x))
    row.names(data) <- NULL
    data <- data.frame(data)
    
    # create the tidy dataframe
    data <- gather(data, key = 'date', value = 'price', 4:15)
    names(data) <- c('food_type', 'type', 'location', 'date', 'price')
    
    data$date <- sapply(data$date, function(x) gsub(pattern = 'X', replacement = '', x))
    data$date <- as.Date(as.integer(data$date), origin = "1899-12-30")
    data <- transform(data, price = as.numeric(price))
    data <- data[complete.cases(data), ]
    data <- data[, c(1,3,4,2,5)]
    return(data)
   
}
```

```r
long <- food_prices

# long <- long[long$food_type != 'Cherries', ]
# long <- long[long$food_type != 'Grapefruit', ]
# long <- long[long$food_type != 'Grapes', ]
# long <- long[long$food_type != 'Artichoke (SF only)', ]

# subset the organic and conventional foods into their own dataframe
organic <- long[long[,'type'] == 'organic', ]
conventional <- long[long[,'type'] == 'conventional', ]

names(organic)[5] <- 'price.organic'
names(conventional)[5] <- 'price.conventional'

organic <- organic[, -c(4)]
conventional <- conventional[, -c(4)]

# merge the two dataframes together to create the wide form of the data
wide <- merge(organic, conventional, by = c('food_type', 'location', 'date', 'category'), all.x = F)

wide$price_diff <- wide$price.organic - wide$price.conventional
wide$percent_diff <- with(wide, (price_diff/price.organic)*100)

wide <- transform(wide, category = factor(category),
                 location = factor(location))
```

# Part Two - Analysis

For analysis, I plotted various forms of the data. Overall, the data needed to be grouped and summarized in different forms. The dplyr package makes this very simple and in the following code this is done. 

The first set of group by actions grouped the data by category of food, either vegetable or fruit, and the year the prices were reported for. The dataframe returns the mean and standard deviation used for plotting.

```r
type_percent <- data.frame(wide %>%
                           group_by(category, year(wide$date)) %>%
                           summarise('mean_percent_premium' = mean(percent_diff),
                                     'std_percent_premium' = sqrt(var(percent_diff))))
```

The following shows what the grouped data looks like after running the above code. 

```r
type_percent
```

The next set of grouping took the above function a step further and grouped it by the month as well to investigate price changes throughout the year.

```r
type_percent_month <- data.frame(wide %>%
                               group_by(category, year(wide$date), month(wide$date)) %>%
                               summarise('mean_percent_premium' = mean(percent_diff),
                                         'std_percent_premium' = sqrt(var(percent_diff))))

type_percent_month$date <- with(type_percent_month, paste(as.character(year.wide.date.), 
                                                          as.character(month.wide.date.), sep = '-'))
type_percent_month$date <- with(type_percent_month, paste0(date, '-1'))
type_percent_month$date <- as.Date(type_percent_month$date, format = '%Y-%m-%d')
```

The following shows what the grouped data looks like after running the above code. 

```r
type_percent_month
```

And finally, the last grouping included location to investigate if there were any price changes by location present in the data.

```r
type_percent_location <- data.frame(wide %>%
                               group_by(category, year(wide$date), location) %>%
                               summarise('mean_percent_premium' = mean(percent_diff),
                                         'std_percent_premium' = sqrt(var(percent_diff))))
```

The following shows what the grouped data looks like after running the above code. 

```r
type_percent_location
```
